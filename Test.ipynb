```python
from keras import backend as K
```

```python
session = K.get_session()
```

```python
feats = K.constant(np.ones(shape=[30,13,13,256]))
```

```python
grid_shape = K.shape(feats)
grid_y = K.arange(0,grid_shape[0])
grid_y = K.expand_dims(grid_y,axis=1)
grid_y = K.tile(grid_y,[1,grid_shape[1]])

grid_x = K.arange(0,grid_shape[1])
grid_x = K.expand_dims(grid_x,axis=0)
grid_x = K.tile(grid_x,[grid_shape[0],1])

grid = K.stack((grid_x,grid_y),axis=-1)
grid = K.expand_dims(grid,axis=2)
```

```python
print(grid)
```

```python
result = session.run(grid)
print(result.shape)
```
